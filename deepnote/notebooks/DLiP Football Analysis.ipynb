# DLiP 2025 Football Analysis

The code below uses the kloppy framework to load data from the Wyscout open data set (Pappalardo et al., 2019). Note that this is only one match for now.

```
from kloppy import wyscout

dataset = wyscout.load_open_data(
    match_id=1694390,
    
    # Optional arguments
    #event_types=["shot", "pass"],
    coordinates="wyscout"
)

df = dataset.to_df()
```

```
df.info()
```
